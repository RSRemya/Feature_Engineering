```python
data.set_index('day', inplace = True)
```

```python
data
```

```python
del_any_incomplete_rec = data.dropna(how='any')
del_any_incomplete_rec
```

```python
del_rec_having_all_miss = data.dropna(how='all')
del_rec_having_all_miss
```

```python
new_data = del_rec_having_all_miss
```

```python
forward_fill_data= new_data.fillna(method='ffill')
forward_fill_data
```

```python
backward_fill_data = new_data.fillna(method='bfill')
backward_fill_data
```

```python
interpolate_data = new_data.interpolate()
interpolate_data
```

```python
interpolate_data_specific= new_data.interpolate(method = 'time')
interpolate_data_specific
```

```python
interpolate_data_specific.to_csv('clean_data.csv')
```
